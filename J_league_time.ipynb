```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option("max_columns", 100)
pd.set_option('max_rows',100)
pd.set_option('max_info_columns', 100)
```

```python
train_1 = pd.read_csv("train_new.csv")
train_2 = train_1.copy()
```

```python
import datetime as dt
time = train_2.loc[:, ["time"]].values.astype('str')
time_tl = []
# time[0][0]

for i in range(len(time)):
# 文字列を時間に変換
    s = dt.datetime.strptime(time[i][0], '%H:%M')
    time_tl.append(s)
train_2["time"] = time_tl
#試合開始時間を30分ごとにまとめて平均
train_2.groupby(pd.Grouper(key='time', freq='1800s')).mean()
```

```python
time = train_2["time"]
time_list = []
a = dt.datetime(1900, 1, 1, 0, 0, 0)
b = dt.datetime(1900, 1, 1, 23, 59, 59)

while a < b:
    time_list.append(a)
    a += dt.timedelta(minutes = 30)

time_list
```

```python
for i in range(len(time)):
    for j in range(len(time_list)): 
        if time_list[j] > time[i]:
            time[i] = j - 1
            break
```
