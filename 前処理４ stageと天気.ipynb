# ステージ

```python
mapping = {"Ｊ１": 0, "Ｊ２":1}
train_new['stage'] = train_new['stage'].map(mapping)
```

# 天気

```python
train_new["weather"] = train_new["weather"].astype("str")
```

```python
#天気ごとに中央値算出
def weather(x):
    if x in ("雨" or "雪"):
        r = train_new[train_new["weather"].str.contains("雨" or "雪")]["y"].median()
    elif x in ("晴"):
        r = train_new[train_new["weather"].str.contains("晴")]["y"].median()
    else:
        r = train_new[train_new["weather"].str.contains("曇" or "屋内")]["y"].median()
    return r
```

```python
train_new["weather"] = train_new["weather"].apply(weather)
```
