```python
#北海道
list_1 = []
def distance_1(a):
#     a = [points_x(経度), points_y]
#     b = [points_x_1(経度), points_y_1]
    b = [141.34694, 43.06417]
    for i  in range(len(a[0])):
        r = 6378 * 2 * math.pi 
        x = (b[0] - a[0][i])
        y = b[1] - a[1][i]
        d = math.sqrt(((x / 360) * r)**2 + ((y / 360) * r)**2)
        #d = r * math.sqrt( x**2 + y**2 ) / 360
        list_1.append(d)
    return list_1

#distance([points_x,points_y],[0.111,3.213213])
#この形で値渡す必要がある。
```
