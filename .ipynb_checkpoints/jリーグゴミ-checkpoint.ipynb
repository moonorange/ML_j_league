```python
#home.away,stadiumをダミー変数化
df1 = pd.get_dummies(train_2['home'])
df2 = pd.get_dummies(train_2['away'])
df3 = pd.get_dummies(train_2['stadium'])
```

```python

#マッピングJ1
train_2['stage'] = train_2['stage'].map( {'Ｊ１': 0, 'Ｊ２': 1} ).astype(int)
    #マッピングgameday
    
train_2["time"] = train_2["time"].str.split(":").str[0].astype(int)
train_2["match"] = train_2["match"].str.split("節").str[0]
train_2["match"] = train_2["match"].str.replace("第","").astype(int) #第を削除する
train_2 = pd.concat([train_2,train_2['gameday'].str.split('(', expand=True)], axis=1).drop('gameday', axis=1)#/以降の文字列を分離。曜日を分離した
train_2 = pd.concat([train_2, df1,df2,df3],axis=1)#さっきダミー変数化した3つのdfを結合
train_2[0] = train_2[0].str.split("/").str[0].astype(int)#日付データの/以下を消去
train_2[1] = train_2[1].str.replace(")","")#分離した曜日データら（を削除。土)
drop_elements = ['id', 'year', 'gameday','home','away','stadium','tv',1]#いらない要素削除
train_2 = train_2.drop(drop_elements,axis=1)
```
