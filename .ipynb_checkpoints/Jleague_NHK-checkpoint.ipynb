```python
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# グラフをインラインで表示させる
%matplotlib inline

# Pandasでの表示が省略されないように設定
pd.set_option("max_columns", 100)
pd.set_option('max_rows',100)
pd.set_option('max_info_columns',100)
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
condition_2 = pd.read_csv("condition_add.csv")
stadium = pd.read_csv("stadium.csv")
add2014 = pd.read_csv("2014_add.csv")
train_add = pd.read_csv("train_add.csv")
train_new = pd.read_csv("train_new.csv")
```

```python
train_new.head()
```

```python
#NHK総合での放送があるものを１、それ以外は０
def NHK(x):
    if 'ＮＨＫ総合' in x:
        r = 1
    else:
        r = 0
    return r
```

```python
train["tv"] = train["tv"].astype('str')
train_new['NHK'] = train_new['tv'].apply(NHK)
```

```python
train_new.head()
```

```python
data = pd.concat([train_new['y'], train_new["NHK"]], axis=1)
f, ax = plt.subplots(figsize=(20, 5))
fig = sns.boxplot(x="NHK", y="y", data=data)
```

```python
train_new.groupby('NHK')['y'].count()
```

```python
def NHK_2(x):
    if 'ＮＨＫ' in x:
        r = 1
    else:
        r = 0
    return r
```

```python
train_new['NHK_2'] = train_new['tv'].apply(NHK_2)
```

```python
train_new.groupby('NHK_2')['y'].count()
```

```python
data = pd.concat([train_new['y'], train_new["NHK_2"]], axis=1)
f, ax = plt.subplots(figsize=(20, 5))
fig = sns.boxplot(x="NHK_2", y="y", data=data)
```
