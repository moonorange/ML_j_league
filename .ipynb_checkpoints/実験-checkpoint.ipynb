```python
# ライブラリの読み込み
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


# グラフをインラインで表示させる
%matplotlib inline

# Pandasでの表示が省略されないように設定
pd.set_option("max_columns", 100)
pd.set_option('max_rows',1000)
pd.set_option('max_info_columns',100)
```

```python
# データの読み込み（csvファイル）
# ヘッダーが無いcsvファイルを読み込むときは引数に'header=None'を指定
train_2 = pd.read_csv('train_new.csv')
test = pd.read_csv('test_new.csv')

```

```python
train_2 = train_2.dropna(subset=["y"])
```

```python
drop_elements = ['home_team','home_01','home_02','home_03',
'home_04',
'home_05',
'home_06',
'home_07',
'home_08',
'home_09',
'home_10',
'home_11',
'away_team',
'away_01',
'away_02',
'away_03',
'away_04',
'away_05',
'away_06',
'away_07',
'away_08',
'away_09',
'away_10',
'away_11',
'referee']
```

```python
train_2= train_2.drop(drop_elements,axis=1)
```

```python


train_2['stage'] = train_2['stage'].map( {'Ｊ１': 0, 'Ｊ２': 1} ).astype(int)

    
train_2["time"] = train_2["time"].str.split(":").str[0].astype(int)
train_2["match"] = train_2["match"].str.split("節").str[0]
train_2["match"] = train_2["match"].str.replace("第","").astype(int) 
train_2 = pd.concat([train_2,train_2['gameday'].str.split('(', expand=True)], axis=1).drop('gameday', axis=1)

train_2[0] = train_2[0].str.split("/").str[0].astype(int)
train_2[1] = train_2[1].str.replace(")","")
drop_elements = ['year',1,'tv']
train_2 = train_2.drop(drop_elements,axis=1)
```

```python
train_2.info()
```

```python
# string label to categorical values
from sklearn.preprocessing import LabelEncoder

for i in range(train_2.shape[1]):
    if train_2.iloc[:,i].dtypes == object:
        lbl = LabelEncoder()
        lbl.fit(list(train_2.iloc[:,i].values))
        train_2.iloc[:,i] = lbl.transform(list(train_2.iloc[:,i].values))
```

```python
train_2.head(10)
```

```python
train_2['game'] = train_2['home'] + train_2['away']
# interaction between the top 2
train_2['point'] = abs(train_2['home_score'] - train_2['away_score'])
```

```python
# ホールド・アウト法によるデータの分割
# (X, y, テストにどの割合を割くか, 分割のしかたを決める乱数生成器)
# テストデータの割合は、test_size引数に指定することで変更可能です（今回は、train:test = 7:3）
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
import xgboost as xgb
# XGブーストを実行
reg_xgb = xgb.XGBRegressor(max_depth=4)
reg_xgb.fit(X_train, y_train)
print('train: %.3f' % reg_xgb.score(X_train, y_train))
print('test : %.3f' % reg_xgb.score(X_test, y_test))
```
