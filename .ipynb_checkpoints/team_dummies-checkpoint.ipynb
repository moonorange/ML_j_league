```python
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# グラフをインラインで表示させる
%matplotlib inline
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
condition_2 = pd.read_csv("condition_add.csv")
stadium = pd.read_csv("stadium.csv")
add2014 = pd.read_csv("2014_add.csv")
train_add = pd.read_csv("train_add.csv")
train_new = pd.read_csv("train_new.csv")
```

```python
hometeam = pd.DataFrame(train.groupby("home")["y"].sum())
n_homegame = pd.DataFrame(train.groupby("home")["id"].count())
mean_home = pd.DataFrame(hometeam["y"] / n_homegame["id"])
mean_home.columns = ["mean"]
mean_home = mean_home.sort_values(["mean"],ascending = False)
```

```python
awayteam = pd.DataFrame(train.groupby("away")["y"].sum())
n_awaygame = pd.DataFrame(train.groupby("away")["id"].count())
mean_away = pd.DataFrame(awayteam["y"] / n_awaygame["id"])
mean_away.columns = ["mean"]
mean_away = mean_away.sort_values(["mean"],ascending = False)
```

```python
mean_home = mean_home.rename( columns = {'home':'team','mean':'mean_home'} )
mean_away = mean_away.rename( columns = {'away':'team','mean':'mean_away'} )
mean_team = pd.merge(mean_home, mean_away)
```

```python
mean_team['mean_all'] = (mean_team['mean_home'] + mean_team['mean_away']) / 2
mean_team.head()
```

```python
mean_team = mean_team.drop(['mean_home','mean_away'],axis = 1 )
mean_team.head()
```

```python
mean_team = mean_team.rename( columns = {'mean_all':'spectators'})
```
